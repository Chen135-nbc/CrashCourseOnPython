```python
#Example1 range(stop)
for x in range(3):
    print(x)
```

```python
#Example2 range(stop)
for x in range(3+1):
    print(x)
```

```python
#Example3 range(start,stop)
for x in range(2,6):
    print(x)
```

```python
#Example4 range(start,stop)
for x in range(5,10+1):
    print(x)
```

```python
#Example5 range(start,stop,step)
for x in range(4,15+1,2):
    print(x)
```

```python
#Example6 range(start,stop,step)
for x in range(2*2,24,3+2):
    print(x)
```

```python
#Example7 range(start,stop,step)
for x in range(10,0,-2):
    print(x)
```

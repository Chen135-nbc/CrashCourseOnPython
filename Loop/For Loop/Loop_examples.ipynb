```python
#Example3
values = [23,52,59,37,48,66]
sum = 0
length = 0
for value in values:
    sum += value
    length += 1
print("Total sum = " + str(sum))
print("Average = " + str(sum / length))
```

```python
#Example4
product = 1
for n in range(1, 10):
    product *= n  #1-9整数相乘
print(product)
```

```python
#Example5
def to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9
for y in range(0,101,10):
    print(y,to_celsius(y))
```

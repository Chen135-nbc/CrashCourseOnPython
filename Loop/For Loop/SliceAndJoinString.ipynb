```python
# How to slice strings:     [startIndex:endIndex]
string1 = "Greetings, Earthlings"
print(string1[0])
print(string1[4:8])
print(string1[11:])
print(string1[:5])
```

```python
# print backwards:  stride argument
string2 = "Electric Engineer"
print(string2[0::2])
print(string2[::-1]) #print backwards
```

```python
# How to join strings:
# 1. use "+"
print("Hello" + " " + "Lucy")

# 2. use ".join()"
greetings = ["Hello","Lucy","Good","Morning"]
print(" ".join(greetings))
```

```python
# How to combine slicing and joining strings:
phoneNum = '2025551212'
area_code = "(" + phoneNum[:3] + ")"
exchange = phoneNum[3:6]
line = phoneNum[-4:]
print(area_code + " " + exchange + "-" + line)
```

```python
# into a function
def format_phone(phonenum):
    area_code1 = "(" + phonenum[:3] + ")"
    exchange1 = phonenum[3:6]
    line1 = phonenum[-4:]
    return area_code1 + " " + exchange1 + "-" + line1
print(format_phone("2025551212"))
```

```python
def convert_seconds(seconds):
    hours = seconds // 3600
    minutes = (seconds - hours * 3600) // 60
    remaining_seconds = seconds - hours * 3600 - minutes * 60
    return hours, minutes, remaining_seconds   #return 多个值 → Python 自动打包成 tuple
result = convert_seconds(5000)
print(result)
type(result)
```

```python
def convert_second(second):
    hours = second // 3600
    minutes = (second - hours * 3600) // 60
    remaining_seconds = second - hours * 3600 - minutes * 60
    return hours, minutes, remaining_seconds
result = convert_second(5000)
hours, minutes, second = result     #用多个变量接收 → Python 自动解包 tuple
print(hours,minutes,second)
```

```python
# 求List中字符串的平均字符数
animals = ["Lion", "Zebra", "Dolphin", "Monkey"]
chars = 0
for animal in animals:
    chars += len(animal)
    
print("Total characters: {}, Average length: {}".format(chars,chars/len(animals)))
```

```python
winners = ["Ashley", "Dylan", "Reese"]
for index, person in enumerate(winners):
    print("{} - {}".format(index + 1, person))
```

```python
def full_emails(people):
    result = []     #用List存最终结果
    for email, name in people:      # 遍历people，同时把每个tuple解包成email和name
        result.append("{} <{}>".format(name, email))
    return result
print(full_emails([("alex@example.com", "Alex Diego"), ("shay@example.com", "Shay Brandt")]))
```

```python
# 给定一个字符串列表，把每个字符串转换为它的大写形式，生成一个新列表。
words = ["apple", "banana", "cherry"]
new_words = [word.upper() for word in words ]
print(new_words)
```

```python
# 给定一个整数列表，只保留 大于 10 的数字，生成一个新列表。
numbers = [3, 15, 8, 22, 7, 30]
new_numbers = [x for x in numbers if x > 10]
print(new_numbers)
```

```python
#（映射 + 条件结合）
nums = [1, 2, 3, 4, 5, 6]
new_nums = [ i ** 2 for i in nums if i % 2 == 0]
print(new_nums)
```
